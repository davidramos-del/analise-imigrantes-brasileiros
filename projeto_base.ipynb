##**Conhecendo os dados**

### Importando os dados

```python
import pandas as pd
```

```python
df = pd.read_csv('imigrantes_canada.csv')
df.info()
```

### Realizando alguns tratamentos

```python
df.set_index('País', inplace=True)
anos = list(map(str, range(1980, 2014)))
anos
```

### Extraindo a série de dados para o Brasil

```python
brasil = df.loc['Brasil', anos]
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
argentina = df.loc['Argentina', anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}

dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina
```

```python
import matplotlib.pyplot as plt
```

## Evolução de imigrantes brasileiros para o Canadá

```python
import seaborn as sns

fig, (ax1, ax2)= plt.subplots(1,2 , figsize = (18,9))
ax1.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
ax1.xaxis.set_major_locator(plt.MultipleLocator(4))
ax1.set(title = "Brasil - Quantidade de Imigrantes por ano de  1980 a 2026", xlabel = "Anoss", ylabel = "Quantidade de Imigrantes")

ax2.plot(dados_argentina['ano'], dados_argentina['imigrantes'], lw=3)
ax2.xaxis.set_major_locator(plt.MultipleLocator(4))
ax2.set(title = "Argentina - Quantidade de Imigrantes por ano de  1980 a 2026", xlabel = "Anoss", ylabel = "Quantidade de Imigrantes")
plt.show()
```

```python
import seaborn as sns
fig, ax = plt.subplots()
sns.barplot(data = dados_brasil, x=dados_brasil['ano'], y=dados_brasil['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(4))
ax.set(title = "Quantidade de Imigrantes por ano de  1980 a 2026", xlabel = "Anoss", ylabel = "Quantidade de Imigrantes")

plt.show()
```
